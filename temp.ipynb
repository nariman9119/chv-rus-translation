```python
from bs4 import BeautifulSoup
data = open('data/chv-rus/chv_rus.xml')
soup = BeautifulSoup(data, 'xml')

```

```python
counter = 500
for tag in soup.find_all("pair"):
    print(tag.rus.text.replace('\n',' ').strip())
    print(tag.chv.text.replace('\n',' ').strip() + '\n')
    counter -= 1
    if counter == 0:
        break
```

```python
from nltk.tokenize import WordPunctTokenizer

def tokenize(src):
    return ' '.join(WordPunctTokenizer().tokenize(src))

```
